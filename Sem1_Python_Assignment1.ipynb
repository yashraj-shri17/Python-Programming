1. What is the distribution of customerages, and how does age relate to product purchase (ProdTaken)?

```python
age_distribution = dataset['Age'].value_counts().sort_index() #This will sort the Age value 
age_distribution
```

```python
age_distribution.plot(kind='bar', figsize=(10, 6), color='skyblue')#this will plot the graph of distribution of age with fig lenght 10 and breadth 6 and color of plot will be skyblue
```

```python
relation_of_age_with_product_purchase = pd.pivot_table(dataset, values='ProdTaken', index='Age', aggfunc=np.sum)#this will make the pivot table of age and the number of product purchase by each age group
relation_of_age_with_product_purchase
```

```python
relation_of_age_with_product_purchase.plot(kind='bar', figsize=(10, 6), color='skyblue'))#this will make the graph of age and the number of product purchase by each age group
```

6. Are there any gender-based preferences for the type of product pitched (ProductPitched)?

7. What is the relationship between the number of trips(number of trips)taken by a customer and their likelihood to purchase a new travel product?

```python
relation_between_the_number_of_trips_and_product_taken=pd.pivot_table(dataset,values='ProdTaken',index='NumberOfTrips',aggfunc='mean')
relation_between_the_number_of_trips_and_product_taken
```

8. Does having a passport(passport)correlate with a higher number of trips taken or a higher likelihood of purchasing a travel product?

```python
higher_number_trips_passport = dataset.groupby('Passport')['NumberOfTrips'].mean()
higher_number_trips_passport
#here it is slight difference in the average number of trips taken between customers who have a passport and those who do not.
```

```python
higher_likelihood_purchase_passport = dataset.groupby('Passport')['ProdTaken'].mean()
higher_likelihood_purchase_passport
#but here we can see it is more differ in average of likelihood purchasing a travel product for having a passport or not having a passport
# so we can say customers who have a passport are more likely to purchase a travel product compared to those who do not have a passport
```

9. How satisfied are customers with the pitch (pitch satisfaction score),and does this satisfaction influence product purchase?

```python
satisfaction_of_customer_with_pitch=dataset.groupby('PitchSatisfactionScore')['CustomerID'].count()
satisfaction_of_customer_with_pitch
#here we can se more number of customers has been satisfied by the pitch because the count of 3,4and 5 is far better
```

```python
checking_satisfaction_of_customer_with_pitch_influenced_product_purchase=dataset.groupby('PitchSatisfactionScore')['ProdTaken'].mean()
checking_satisfaction_of_customer_with_pitch_influenced_product_purchase
#here we can see  customers with Pitch Satisfaction Scores of 3 and 5 have the highest likelihood of purchasing a product, with proportions of 21.38% and 21.65% respectively.
```

12. Impact of marital status on travel preferences: does marital status affect the choice of travel product, and if so, how do different marital statuses correlate with product purchases (prod taken)?
